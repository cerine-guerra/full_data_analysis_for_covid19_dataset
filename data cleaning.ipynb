# 1. Understanding the data

**General Overview**

This dataset provides a comprehensive view of the COVID-19 pandemic using multiple CSV files, each capturing the data from a different perspective and level of detail. It includes day-to-day records of confirmed cases, deaths, and recoveries at both global and country levels, as well as more granular data for specific regions such as U.S. counties and administrative divisions. Some files focus on historical time-series data, allowing analysis of the evolution of the pandemic over time, while others provide the latest summarized statistics at the country level. Additionally, the dataset incorporates population-based metrics sourced from Worldometer, enabling comparative and normalized analysis between countries. Overall, the dataset supports temporal, geographical, and comparative analysis of COVID-19 trends worldwide.

# **3. Data preparation**

1\ full_grouped.csv


1.   Converting columns to appropriate data types

```python
df_full.dtypes
```

```python
df_full['Date'] = pd.to_datetime(df_full['Date'])
df_full['WHO Region'] = df_full['WHO Region'].astype('category')
num_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases','New deaths','New recovered']
df_full[num_cols] = df_full[num_cols].astype(int)
```

```python
df_full.dtypes
```


2.   creating a meaningful feature

We create a new feature called Case Fatality Rate (CFR), calculated as Deaths / Confirmed for each country on each date. This feature provides a normalized measure of disease severity, allowing comparisons across countries and over time, regardless of total case counts.



```python
df_full['CFR'] = df_full.apply(lambda row: row['Deaths']/row['Confirmed'] if row['Confirmed'] > 0 else 0, axis=1)
```

```python
df_full[['Country/Region', 'Date', 'Confirmed', 'Deaths', 'CFR']].head()
```

# **5. visualizations**

### Visualization 1 – Global Trend Over Time

```python
daily_totals = df_day.groupby('Date')[['Confirmed','Deaths','Recovered']].sum()

plt.figure(figsize=(14,6))
plt.plot(daily_totals.index, daily_totals['Confirmed'], label='Confirmed', color='blue', linewidth=2)
plt.plot(daily_totals.index, daily_totals['Deaths'], label='Deaths', color='red', linewidth=2)
plt.plot(daily_totals.index, daily_totals['Recovered'], label='Recovered', color='green', linewidth=2)
plt.title('Global COVID-19 Progression Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

This line chart shows how confirmed cases, deaths, and recoveries evolved globally over time.
It highlights pandemic growth patterns and helps identify peaks and recovery trends.

### Visualization 2 – Top 10 Countries by Total Confirmed Cases

```python
top10_countries = df_full.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top10_countries.plot(kind='bar', color='pink')
plt.title('Top 10 Countries by Total Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar chart Highlights the 10 countries with the highest total confirmed cases. showing the US with the highest number of confirmed cases where as South Africa , Mexico , Peru , Chile , United Kingdom , Iran have a much lower number at approximetly the same level.
Which allows for an easy comparison of countries most affected by COVID-19.

### Visualization 3 – Average Case Fatality Rate by WHO Region

```python
df_clean['CFR'] = df_clean.apply(lambda row: row['Deaths']/row['Confirmed'] if row['Confirmed']>0 else 0, axis=1)
region_cfr = df_clean.groupby('WHO Region')['CFR'].mean()

plt.figure(figsize=(8,8))
plt.pie(region_cfr, labels=region_cfr.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Average Case Fatality Rate by WHO Region')
plt.tight_layout()
plt.show()

```

This pie chart shows the distribution of average case fatality rate across WHO regions.
It provides a quick sense of which regions experienced higher mortality relative to total cases like America , Eastern Mediterranean and Europe.

### Visualization 4 – U.S. States by Total Confirmed Cases

```python
confirmed_by_state = df_usa.groupby('Province_State')['Confirmed'].sum()

plt.figure(figsize=(14,6))
plt.scatter(confirmed_by_state.index, confirmed_by_state.values, color='red', s=100)
plt.title('Total Confirmed COVID-19 Cases by U.S. State')
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This scatter plot visualizes the total confirmed COVID-19 cases for each U.S. state.
Each dot represents a state, providing a clear overview of the distribution of cases. showing how New York had the highest number of cases compared to the rest of the states .


### Visualization 5 – Top 10 Countries by Tests per Million

```python
top_testing = df_world.sort_values('Tests/1M pop', ascending=False).head(10).set_index('Country/Region')

plt.figure(figsize=(12,6))
top_testing['Tests/1M pop'].plot(kind='barh', color='purple')
plt.title('Top 10 Countries by COVID-19 Tests per Million Population')
plt.xlabel('Tests per Million Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

The horizontal bar chart above  displays countries with the highest COVID-19 testing coverage per million population. where we can tellthat Luxembourg and Monaco had the highest testing coverange making it easier to compare.

# 6. Insights and Conclusions

1. Global trends
---


*  The global COVID-19 progression shows a steady rise in confirmed cases over
*  time, with clear peaks corresponding to major waves.
*  Deaths generally follow confirmed cases with a slight delay, while recoveries increase steadily as the pandemic progresses.
*  The line plot of daily totals clearly shows how the pandemic evolved globally.







2. Country-level differences
---


*  A small number of countries, such as USA, India, and Brazil, account for the highest total confirmed cases, indicating a highly uneven global impact.
*  Countries with the highest tests per million population tend to have better detection of cases, which may explain differences in reported confirmed cases.
*  The Worldometer dataset shows that testing coverage varies significantly, even among countries with similar populations.


3. Regional patterns
---


*  WHO Regions differ in case fatality rates (CFR), with some regions showing higher average mortality relative to total confirmed cases.
*  The pie chart of CFR by WHO Region highlights that Africa and Eastern Mediterranean regions experienced relatively higher mortality in proportion to cases, while other regions had lower CFRs.
*  Confirmed cases per WHO Region show that the Americas and Europe had the highest case counts overall.


4. U.S. county/state insights
---


*  The U.S. states with the highest confirmed cases are concentrated in larger, more populated states such as New York, California, and Texas.
*  County-level data shows localized hotspots, demonstrating that within-state variations are significant.
*  Scatter plots of U.S. states help visualize which states bear the highest burden.


5. Surprising or interesting patterns
---


*  Some countries with high testing rates also report high case numbers, suggesting that detection bias plays a role.
*  Recovery rates are generally high globally, but active cases remain elevated in countries with recent surges.
*  Daily new cases show clear wave-like patterns, indicating recurring peaks rather than a single exponential curve.
*  Despite high case numbers, some regions maintain low mortality, showing potential differences in healthcare response and population demographics.







# Summary and conclution:

The COVID-19 pandemic shows uneven impact across countries, regions, and states, with clear global waves of infection over time. High-population areas and countries with extensive testing report the most cases, while mortality rates vary significantly between regions. These datasets together highlight patterns in confirmed cases, deaths, recoveries, and testing, providing a clear overview of the pandemic's progression and its regional differences.
